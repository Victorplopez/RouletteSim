```python
ax = jamesBond1.plot(y="total", figsize=(15,10), color='red')
jamesBond2.plot(ax=ax, color='red')
jamesBond3.plot(ax=ax, color='red')
jamesBond4.plot(ax=ax, color='red')
jamesBond5.plot(ax=ax, color='red')

jamesBond1a.plot(ax=ax, color='blue')
jamesBond2a.plot(ax=ax, color='blue')
jamesBond3a.plot(ax=ax, color='blue')
jamesBond4a.plot(ax=ax, color='blue')
jamesBond5a.plot(ax=ax, color='blue')

jamesBond1z.plot(ax=ax, color='black')
jamesBond2z.plot(ax=ax, color='black')
jamesBond3z.plot(ax=ax, color='black')
jamesBond4z.plot(ax=ax, color='black')
jamesBond5z.plot(ax=ax, color='black')

ax.set(xlabel='Amount of runs', ylabel='Money',
       title='James Bond 10000 spins')

ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed')
ax.xaxis.grid(color='gray', linestyle='dashed')

L=plt.legend()


for i in range(5):
    L.get_texts()[i].set_text('European')
for i in range(5,10):
    L.get_texts()[i].set_text('American')
for i in range(10,15):
    L.get_texts()[i].set_text('No Zero')

    
ax.axhline(y=1000,color="black", linestyle='--')
plt.show()
```
