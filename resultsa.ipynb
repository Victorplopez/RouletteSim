```python
jamesBond1 = pd.read_csv('results/007/100spins/results1a.csv')
jamesBond2 = pd.read_csv('results/007/100spins/results2a.csv')
jamesBond3 = pd.read_csv('results/007/100spins/results3a.csv')
jamesBond4 = pd.read_csv('results/007/100spins/results4a.csv')
jamesBond5 = pd.read_csv('results/007/100spins/results5a.csv')
```

```python
del jamesBond1["run"]
del jamesBond2["run"]
del jamesBond3["run"]
del jamesBond4["run"]
del jamesBond5["run"]
```

```python
x = [(i+1)*10 for i in range(11)]
ax = jamesBond1.plot(y="total", figsize=(15,10))
plt.xticks(np.arange(min(x), max(x), 10))
jamesBond2.plot(ax=ax)
jamesBond3.plot(ax=ax)
jamesBond4.plot(ax=ax)
jamesBond5.plot(ax=ax)

ax.set(xlabel='Amount of runs', ylabel='Money',
       title='James Bond 100 spins')

ax.set_axisbelow(True)
ax.yaxis.grid(color='gray', linestyle='dashed')
ax.xaxis.grid(color='gray', linestyle='dashed')

L=plt.legend()

for i in range(5):
    L.get_texts()[i].set_text('Run '+ str(i+1))
    
ax.axhline(y=1000,color="black", linestyle='--')
plt.show()
```

```python
display(jamesBond1.describe())
display(jamesBond2.describe())
display(jamesBond3.describe())
display(jamesBond4.describe())
display(jamesBond5.describe())
```
